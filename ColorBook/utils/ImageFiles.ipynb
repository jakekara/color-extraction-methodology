# read_img(path, resize=1):

There are a lot of ways between cv2, PIL and numpy to load an image. This function will abstract that so we do it the same every time. This approach reads an image from disk, conver it to RGBA and return a numpy array. 


```python
def read_img(path, scale=1):
    """
        Read a file from disk and convert from BGR to RGBA
    """

    ret = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGBA)

    if scale != 1:
        ret = cv2.resize(
            ret,
            (int(np.array(ret).shape[1] * scale),
            int(np.array(ret).shape[0] * scale))
        )
    return np.array(ret)
```

```python
# :: ignore-cell ::
from PIL import Image
Image.fromarray(read_img("../../images/jerusalem.mpi.p22-51.100.jpg", scale=0.5))
```
